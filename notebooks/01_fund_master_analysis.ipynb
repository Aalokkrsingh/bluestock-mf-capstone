#### Step 5: AMFI Code Validation (Task 7)

```python
# Load master and historical tracking datasets
fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")

# Extract AMFI codes as sets for fast cross-referencing
fund_codes = set(fund_master['amfi_code'])
nav_codes = set(nav_history['amfi_code'])

# Identify AMFI codes present in master list but missing from NAV history
missing_codes = fund_codes - nav_codes

print("Missing Codes:")
print(missing_codes)
```
